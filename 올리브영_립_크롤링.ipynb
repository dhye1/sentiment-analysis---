```python
from bs4 import BeautifulSoup
from selenium import webdriver 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import pandas as pd
from tqdm.notebook import tqdm
import time
import re
import sys
from selenium.webdriver.common.keys import Keys #키보드 키를 제어하는 라이브러리
keys = Keys()
```

```python
url = 'https://www.oliveyoung.co.kr/store/goods/getGoodsDetail.do?goodsNo=A000000121562&dispCatNo=100000100020006&trackingCd=Cat100000100020006_Small'
driver = webdriver.Chrome("C:/py_data/chromedriver.exe")
driver.get(url)
time.sleep(2)
```

```python
key = driver.find_element_by_css_selector('#reviewInfo > a') #리뷰페이지로 이동
key.click()
driver.execute_script("window.scrollTo(0, document.body.scrollHeight-50);")
time.sleep(3)
#화면 맨 아래로 스크롤 후 기다림
date = str(input("저장날짜를 입력하세요: "))
stop = int(input("몇 세트를 크롤링 할까요?(1-11페이지가 한세트): "))
```

```python
# 이미 존재하는 버튼 클릭
color = driver.find_element_by_class_name('prd_option_box.box_select').click()

# 추가로 버튼이 나타날 때까지 기다리기
wait = WebDriverWait(driver, 10) # 10초 동안 기다리는 인스턴스 선언

# 추가 버튼 클릭(색상 설정)
#만찢코랄 
driver.find_element_by_css_selector('#gdasContentsArea > div > div.prd_option_box.box_select.open > ul > li:nth-child(6)').click()
#미모열일
#driver.find_element_by_css_selector('#gdasContentsArea > div > div.prd_option_box.box_select.open > ul > li:nth-child(21)').click()
```

```python
review_lst = []
page_btn = ['strong','a:nth-child(2)', 'a:nth-child(3)', 'a:nth-child(4)', 'a:nth-child(5)', 'a:nth-child(6)',
           'a:nth-child(7)','a:nth-child(8)','a:nth-child(9)','a:nth-child(10)','a.next']
page_btn2 = ['strong','a:nth-child(3)', 'a:nth-child(4)', 'a:nth-child(5)', 'a:nth-child(6)',
           'a:nth-child(7)','a:nth-child(8)','a:nth-child(9)','a:nth-child(10)', 'a:nth-child(11)','a.next']

count = 0
while count < stop:
    if count < 1:
        for pagenum in page_btn:
            page= driver.find_element_by_css_selector('#gdasContentsArea > div > div.pageing > '+str(pagenum)+'')
            page.click()
            for i in range(1,10):
                    html = driver.page_source
                    soup = BeautifulSoup(html, "html.parser")
                    review_all = soup.find_all('div',class_='txt_inner')
                    review = review_all[i].text
                    rv=re.sub('[^0-9ㄱ-ㅣ가-힣 ]', " ",review)
                    review_lst.append(rv)
    else:
        for pagenum in page_btn2:
            page= driver.find_element_by_css_selector('#gdasContentsArea > div > div.pageing > '+str(pagenum)+'')
            page.click()
            for i in range(1,10):
                    html = driver.page_source
                    soup = BeautifulSoup(html, "html.parser")
                    review_all = soup.find_all('div',class_='txt_inner')
                    review = review_all[i].text
                    rv=re.sub('[^0-9ㄱ-ㅣ가-힣 ]', " ",review)
                    review_lst.append(rv)        
    count = count + 1
            
olive = pd.DataFrame({'리뷰':review_lst})
olive.to_csv("c:/py_data/올리브영 리뷰" +date+ '.csv',
               encoding='cp949')
review_lst
```
